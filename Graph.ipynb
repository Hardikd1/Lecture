```python
d3 = sns.load_dataset("tips")
```

```python
d3.head()
```

```python
d3[['total_bill','tip']].iplot(x='total_bill',y = 'tip',mode='markers',size = 5,xTitle = 'Bill amt',yTitle='Tip')
```

```python
# plot he data in the 3d format

d3.iplot(kind='scatter3d',x='total_bill',y='tip',z='size')
```

```python
d3.iplot(kind='bubble3d',x='total_bill',y='tip',z='size',size = "size")
```

```python
# we can do the same ploting using seaborn as well

sns.relplot(x='total_bill',y='tip',data=d3)
```
